### Prediction between HandlingTime and Downtime

```python
from sklearn.preprocessing import MinMaxScaler
data = pd.read_csv(r"C:\Users\15878\DA\SGD_Data_Pred.csv")
# create a MinMaxScaler object
scaler = MinMaxScaler()

# select columns to scale
cols_to_scale = ['HT']
cols_to_scale1 = ['Down_Time']

# apply Min-Max scaling to selected columns
data[cols_to_scale] = scaler.fit_transform(data[cols_to_scale])
data[cols_to_scale1] = scaler.fit_transform(data[cols_to_scale1])
print(data[cols_to_scale],data[cols_to_scale1])
```

```python
# Split data into features and target variable
X = data[cols_to_scale]
y = data[cols_to_scale1]
print(X,y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.7,test_size=0.3, random_state=42)
```

```python
# Create a Decision Tree Regressor
dtr = DecisionTreeRegressor(max_depth=5, random_state=42)
```

```python
# Train the regressor on the training data
dtr.fit(X_train, y_train)
```

```python
y_pred = dtr.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```
